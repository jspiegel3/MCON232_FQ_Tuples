```python
def write_file():
  with open("numbers.txt", "w") as f :
    f.write("1,2,3\n4,5,6\n7,8,9\n")

def read_file1():
  di_list = []
  with open("numbers.txt", "r") as f:
    for row in f:
      di_list.append(list(row.replace(',', '').replace("\n", '')))

  return di_list

def read_file_csv():
  di_list = []

  new_row = []
  with open("numbers.txt", "r") as f:
    for row in f :
      new_row = []
      for char in row:
        if char not in ",\n":
          new_row.append(char)
      di_list.append(new_row)
  return di_list


print(read_file_csv())
```
